```python
covid_DF['Active case'] = covid_DF['Confirmed'] - (covid_DF ['Cured'] + covid_DF ['Deaths'])
covid_DF .tail()
```

```python
#What is the trend of COVID-19 confirmed cases over time for a specific state?
```

```python

# Assuming covid_DF is already loaded as a DataFrame
state = 'Maharashtra'  # Example state
state_data = covid_DF[covid_DF['State/UnionTerritory'] == state]

plt.figure(figsize=(12, 6))
plt.plot(state_data['Date'], state_data['Confirmed'], label='Confirmed Cases')
plt.title(f'Trend of Confirmed COVID-19 Cases Over Time in {state}')
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
  
    
#This code visualizes the trend of confirmed COVID-19 cases over time for a specific state.
#It ensures the Date column is in the correct datetime format and filters the data for the chosen state.
```

```python
# How do the number of cured cases compare across different states on a specific date?
```

```python
specific_date = '2021-05-01'  # Example date
date_data = covid_DF[covid_DF['Date'] == specific_date]

plt.figure(figsize=(12, 8))
plt.barh(date_data['State/UnionTerritory'], date_data['Cured'], color='green')
plt.title(f'Cured Cases Across Different States on {specific_date}')
plt.xlabel('Number of Cured Cases')
plt.ylabel('State/UnionTerritory')
plt.tight_layout()
plt.show()


#This code visualizes the number of cured COVID-19 cases across different states for a specific date.
#It filters the data for the chosen date and creates a horizontal bar chart.
```

```python
# What is the death rate (deaths per confirmed cases) over time for a specific state?
```

```python
state = 'Delhi'  # Example state
state_data = covid_DF[covid_DF['State/UnionTerritory'] == state]
state_data['DeathRate'] = state_data['Deaths'] / state_data['Confirmed'] * 100

plt.figure(figsize=(12, 6))
plt.plot(state_data['Date'], state_data['DeathRate'], color='red', label='Death Rate')
plt.title(f'Death Rate Over Time in {state}')
plt.xlabel('Date')
plt.ylabel('Death Rate (%)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#This code visualizes the death rate (deaths per confirmed cases) of COVID-19 over time for a specific state. 
#It calculates the death rate and plots it against the date for the chosen state.

```

```python
#How do the number of confirmed cases differ across various states/union territories as of the latest date in the dataset?
```

```python
latest_date = covid_DF['Date'].max()
latest_data = covid_DF[covid_DF['Date'] == latest_date]

plt.figure(figsize=(14, 10))
plt.bar(latest_data['State/UnionTerritory'], latest_data['Confirmed'], color='blue')
plt.title(f'Confirmed COVID-19 Cases Across States/Union Territories as of {latest_date}')
plt.xlabel('State/UnionTerritory')
plt.ylabel('Number of Confirmed Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

#This code visualizes the number of confirmed COVID-19 cases across different states/union territories as of the latest date.
#It filters the data for the latest date in the dataset and creates a bar chart to display the confirmed cases.
```

```python
#What is the distribution of the number of deaths across all states/union territories as of the latest date?
```

```python
latest_date = covid_DF['Date'].max()
latest_data = covid_DF[covid_DF['Date'] == latest_date]

plt.figure(figsize=(12, 8))
plt.hist(latest_data['Deaths'], bins=20, color='purple', edgecolor='black')
plt.title(f'Distribution of COVID-19 Deaths Across States/Union Territories as of {latest_date}')
plt.xlabel('Number of Deaths')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

#This code visualizes the distribution of COVID-19 deaths across different states/union territories as of the latest date.
#It filters the data for the latest date in the dataset and creates a histogram to display the distribution of deaths.
```

```python
latest_date = covid_DF['Date'].max()
latest_data = covid_DF[covid_DF['Date'] == latest_date]

plt.figure(figsize=(14, 10))
plt.pie(latest_data['Confirmed'], labels=latest_data['State/UnionTerritory'], autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title(f'Confirmed COVID-19 Cases Distribution Across States/Union Territories as of {latest_date}')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()

#higst stast maharashtra =21.3
```
